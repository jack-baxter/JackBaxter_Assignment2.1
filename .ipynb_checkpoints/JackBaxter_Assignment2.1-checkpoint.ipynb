```python
!aws s3 cp s3://sagemaker-us-east-1-883727203134/dataset.csv \
           s3://sagemaker-us-east-1-883727203134/table/dataset.csv
```

```python
#sql table creation
statement = """CREATE EXTERNAL TABLE IF NOT EXISTS {}.{}(
         Unnamed int,
         track_id string,
         artists string,
         album_name string,
         track_name string,
         popularity int,
         duration_ms int,
         explicit boolean,
         danceability float,
         energy float,
         key int,
         loudness float,
         mode int,
         speechiness float,
         acousticness float,
         instrumentalness float,
         liveness float,
         valence float,
         tempo float,
         time_signature int,
         track_genre string
) ROW FORMAT DELIMITED FIELDS TERMINATED BY '\\,' LINES TERMINATED BY '\\n' LOCATION '{}'
TBLPROPERTIES ('compressionType'='gzip', 'skip.header.line.count'='1')""".format(
    database_name, table_name_csv, s3_private_path_csv
)
print(statement)
```

```python
import pandas as pd
pd.read_sql(statement, conn)
```

```python
statement = "SHOW TABLES in {}".format(database_name)
df_show = pd.read_sql(statement, conn)
df_show.head(5)
```

```python
if table_name_csv in df_show.values:
    ingest_create_athena_table_csv_passed = True
%store ingest_create_athena_table_csv_passed
%store
```

```python
#HW Problem 1
#List artist, track_name, and popularity for songs that have a popularity greater than or equal to 99
statement = """SELECT artists, track_name, popularity FROM {}.{}
    WHERE popularity >= 99""".format(
    database_name, table_name_csv
)
print(statement)
df = pd.read_sql(statement, conn)
df.head(5)
```

```python
#HW Problem 2
#List artists with an average popularity of 92
statement = """SELECT artists FROM {}.{}
    WHERE popularity >= 99""".format(
    database_name, table_name_csv
)
print(statement)
df = pd.read_sql(statement, conn)
df.head(5)
```

```python
#HW Problem 3 
#List the Top 10 genres with the highest average energy
statement = """SELECT 
    track_genre,
    ROUND(AVG(energy), 4) AS average
FROM {}.{}
GROUP BY track_genre
ORDER BY average DESC
LIMIT 10""".format(
    database_name, table_name_csv
)
print(statement)
df = pd.read_sql(statement, conn)
df.head(5)
```

```python
#HW Problem 4
#How many tracks is Bad Bunny on?
statement = """SELECT 
    COUNT(*) AS track_count
FROM {}.{}
WHERE artists = 'Bad Bunny'""".format(
    database_name, table_name_csv
)
print(statement)
df = pd.read_sql(statement, conn)
df.head(5)
```

```python
#HW Problem 5
#Show the top 10 genres in terms of popularity, sorted by their most popular track
statement = """SELECT 
    track_genre,
    MAX(popularity) AS max_popularity
FROM {}.{}
GROUP BY track_genre
ORDER BY max_popularity DESC
LIMIT 10""".format(database_name, table_name_csv)
print(statement)
df = pd.read_sql(statement, conn)
df.head(5)
```
